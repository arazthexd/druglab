```python
from rdkit import Chem
from druglab.io.readers import CSVFormatReader
from druglab.db import MoleculeTable
from druglab.pipe import (
    MoleculeDesalter, MoleculeSanitizer, 
    ElementFilter, MorganFeaturizer, MACCSFeaturizer
)
from druglab.pipe import Pipeline
```

```python
recs = CSVFormatReader(
    smiles_col="Smiles", 
    name_col="Molecule ChEMBL ID", 
    delimiter=';',
    on_error='skip'
).read_all('data/ca2_inhibitors.csv')

recs[:5]
```

```python
table = MoleculeTable.from_records(recs)

table
```

```python
pipe = Pipeline([
    MoleculeDesalter(),
    MoleculeSanitizer(),
    ElementFilter(),
    MorganFeaturizer(),
    MACCSFeaturizer()
])

from rdkit import rdBase
rdBase.DisableLog('rdApp.*')

out = pipe.run(table)
out
```

```python
out.history
```

```python
out.save("./out_simple_pipeline")
```
